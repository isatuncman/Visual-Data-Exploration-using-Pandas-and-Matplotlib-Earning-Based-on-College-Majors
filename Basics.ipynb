## Scatter Matrices

```python
from pandas.plotting import scatter_matrix
```

### 1) Sample Size vs Median

```python
scatter_matrix(recent_grads[['Sample_size', 'Median']])
plt.show()
```

We can see that popular majors make moderate amount of money and the highest salaries belong to unpopular majors.

### 2) Sample Size, Median, Unemployment Rate

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize = (10,10))
plt.show()
```

Here, we can see that Unemployment rate distribution is a bit right skewed and high-salary-majors have smaller unemployment rate.

## Bar Plots

### 1) Women Share of Majors:  Minimum 10

```python
recent_grads.sort_values('ShareWomen')[:10].plot.bar(x = 'Major', y= 'ShareWomen', legend = False)
plt.show()
```

### 2) Women Share of Majors:  Max 10

```python
recent_grads.sort_values('ShareWomen')[-10:].plot.bar(x = 'Major', y= 'ShareWomen', legend = False)
plt.show()
```

### 3) Unemployment rate of Majors: Minimum 10

```python
recent_grads.sort_values('Unemployment_rate')[:10].plot.bar(x = 'Major', y ='Unemployment_rate', legend = False)
plt.show()
```

```python
recent_grads.sort_values('Unemployment_rate')[-10:].plot.bar(x = 'Major', y ='Unemployment_rate', legend = False)
plt.show()
```

## Grouped Bar Plots

### 1) Men vs Women Number: Men's Top 10 Majors

```python
recent_grads.sort_values('Men', ascending = False)[['Men', 'Women', 'Major']][:10].plot.bar(x = 'Major')
plt.show()
```

### 2) Men vs Women Number: Men's Minimum 10 Majors

```python
recent_grads.sort_values('Men')[['Men', 'Women', 'Major']][:10].plot.bar(x = 'Major')
plt.show()
```

## Stacked Bar Plots

### Men and Women distribution of the Majors with Top Median Salary: 

```python
recent_grads.sort_values('Median', ascending = False)[['Men','Women','Major']][:10].plot.bar(x = 'Major', stacked = True)
plt.show()
```

Chemical and Computer enginnering are the majors with large population and significant women number.
